# Working with Text Data

```python
import pandas as pd
chicago=pd.read_csv("practice_csv\chicago.csv").dropna(how="all")
chicago["Department"]=chicago["Department"].astype("category")
chicago["Position Title"]=chicago["Position Title"].astype("category")
chicago
```

```python
chicago.info()
```

```python
chicago["Name"].str.lower().head(3) # to convert column to lower case 
```

```python
chicago["Name"].str.lower().str.upper().head(3)
```

```python
chicago["Name"].str.lower().str.title().head(3)# to convert column to title case (first letter of each word as capital)
```

```python
chicago["Name"].str.len().head(3)# to get length of each string
```

```python
chicago["Department"]=chicago["Department"].str.replace("MGMNT","MANAGEMENT").astype("category")
chicago.head()
```

```python
chicago["Employee Annual Salary"]=chicago["Employee Annual Salary"].str.replace("$","").astype("float")
chicago.head()
```

```python
chicago["Employee Annual Salary"].sum()
chicago["Employee Annual Salary"].mean()
chicago["Employee Annual Salary"].std()
chicago["Employee Annual Salary"].nlargest(3)
```

```python
chicago["Name"].str.strip().head(3) #remove leding and trailing whitespace 
```

```python
#chanaging dataframe index to Name column
chicago.set_index("Name",inplace=True)
chicago.head(3)
```

```python
chicago.index=chicago.index.str.lower()#these method can be used with index method
chicago.head(3)
```

```python
chicago.columns=chicago.columns.str.upper()
chicago.head(3)
```

# 4.Filtering with String Methods

```python
cond=chicago["POSITION TITLE"].str.lower().str.contains("water")#looks for "water" anywhere in the string
chicago[cond].head(3)
```

```python
cond=chicago["POSITION TITLE"].str.lower().str.startswith("water")#looks for "water" only at the beginning 
chicago[cond].head(3)
```

```python
cond=chicago["POSITION TITLE"].str.lower().str.endswith("ist")#looks for only only at the end of the string 
chicago[cond].head(3)
```
