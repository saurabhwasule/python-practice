# 4.Retrieve Rows by Index postion with .iloc[ ]

```python
# Single selections using iloc and DataFram
bond.iloc[1] # Returns panda series with index 1
```

```python
# Rows:
bond.iloc[0] # first row of bond frame 
bond.iloc[1] # second row of bond frame 
bond.iloc[-1] # last row of bond frame
```

```python
# Columns:
bond.iloc[:,0] # first column of bond frame (Actor)
bond.iloc[:,1] # second column of bond frame (Director)
bond.iloc[:,-1] # last column of bond frame (Salary)
```

```python
# Multiple row and column selections using iloc and bondFrame
bond.iloc[0:5] # first five rows of bondframe
bond.iloc[:, 0:2] # first two columns of bond frame with all rows
bond.iloc[[0,3,6,20], [0,2,4]] # 1st, 4th, 7th, 25th row + 1st 2nd 4th columns.
```

```python
bond.iloc[0:5, 0:2] # first 5 rows and 1st, 2th, 3rd columns of bond frame 
```

# 13.Filtering with the .where() and query() Method

```python
#Earlier we have seen below way
cond1=bond["Actor"]=="Sean Connery"
cond2=bond["Box Office"]>800
bond[cond1 & cond2]
```

```python
#keep original data frame but isolated the rows satisfying  condition
# where() method to return a modified DataFrame that holds NaN values for all rows that don't match our provided condition.
bond.where(cond1 & cond2)
```

```python
# to use query method no space should be present in header.Replacing space with underscore
bond.columns=[column_name.replace(" ","_") for column_name in bond.columns]
bond.head(1)
```

```python
# it faster than normal way
bond.query('Actor =="Sean Connery" and Box_Office > 800')
```

```python
bond.query("Actor in['Sean Connery','Roger Moore','David Niven'] and Director not in ['Guy Hamilton']")
```

# .apply() Method

```python
#  .apply() method on a Series to apply a Python function on every value within it.
def add_million(num):
    if num!=0:
        return str(num)+" Million !"
    else:
        return str(num)
# bond["Box_Office"]=bond["Box_Office"].apply(add_million)
```

```python
# applying same logic to all float columns
bond["Box_Office"].fillna(0,inplace=True)
bond["Budget"].fillna(0,inplace=True)
bond["Bond_Actor_Salary"].fillna(0,inplace=True)
# for col in ["Box_Office","Budget","Bond_Actor_Salary"]:
#         bond[col]=bond[col].apply(add_million)
```

```python
bond.head(10)
```

```python
# The .apply() method applies a Python function on a row-by-row basis in a DataFrame.
def good_movie(row):
    actor=row[1]
    budget=row[4]
    if actor=="Sean Connery" and budget>40:
        return "The Best"
    elif actor=="Roger Moore" and budget>50:
        return "Enjoyable"
    else:
        return "No Clue"
```

```python
bond["comment"]=bond.apply(good_movie,axis="columns")#add comment based on comments from above 
bond.head(5)
```
