# 3.Retrieve Rows by Index Label with .loc[ ]

```python
# to retrieve the row in faster way from million of rows we use loc instead of normal way
bond.reset_index(drop=True,inplace=True)
bond.set_index("Film",inplace=True)
bond.head(3)
```

```python
bond.loc["Goldfinger"] # Returns panda series if single row is return
```

```python
bond.loc["Casino Royale"]  # Returns panda dataframe if more than one row is return
```

```python
# bond.loc["Golden Bond"]  # Returns error it movie name is not present
```

```python
bond.loc["You Only Live Twice":"Live and Let Die"]  # Returns panda dataframe from You Only Live Twice to Live and Let Die
```

```python
bond.loc["You Only Live Twice":,["Actor"]]  # Returns panda dataframe from You Only Live Twice to end
```

```python
bond.loc[:"Diamonds Are Forever",["Actor","Director"]]  # Returns panda dataframe from starting to Diamonds Are Forever with Actor and director column
```

```python
bond.loc[["You Only Live Twice","Live and Let Die","Golden Bond"]]  # Search more than one movie and one of the movie in not in present
#As Golden Bond is not present it will add null row against the index and will be raising keyerror in future 
```

```python
# To check movie is present or not we can use in operator
"Golden Bond" in bond.index
```

# 4.Retrieve Rows by Index postion with .iloc[ ]

```python
# Single selections using iloc and DataFram
bond.iloc[1] # Returns panda series with index 1
```

```python
# Rows:
bond.iloc[0] # first row of bond frame 
bond.iloc[1] # second row of bond frame 
bond.iloc[-1] # last row of bond frame
```

```python
# Columns:
bond.iloc[:,0] # first column of bond frame (Actor)
bond.iloc[:,1] # second column of bond frame (Director)
bond.iloc[:,-1] # last column of bond frame (Salary)
```

```python
# Multiple row and column selections using iloc and bondFrame
bond.iloc[0:5] # first five rows of bondframe
bond.iloc[:, 0:2] # first two columns of bond frame with all rows
bond.iloc[[0,3,6,20], [0,2,4]] # 1st, 4th, 7th, 25th row + 1st 2nd 4th columns.
```

```python
bond.iloc[0:5, 0:2] # first 5 rows and 1st, 2th, 3rd columns of bond frame 
```
