# 1.This Module's Dataset + Memory Optimization

```python
import pandas as pd
emp=pd.read_csv("practice_csv/employees.csv",parse_dates=["Start Date","Last Login Time"])
# optimize data frame performance by reducing size
emp["Gender"]=emp["Gender"].astype("category")
emp["Senior Management"]=emp["Senior Management"].astype("bool")
emp["Team"]=emp["Team"].astype("category")
# # convert start datte and last login Time to datetime. We can parse date in read csv instead of two line
# emp["Start Date"]=pd.to_datetime(emp["Start Date"])
# emp["Last Login Time"]=pd.to_datetime(emp["Last Login Time"]) # date time current date will be appended,we can ignore it for now
emp
```

# 2.Filter a DataFrame Based on A Condition

```python
#get all rows where Gender is Male
cond_gender=emp["Gender"]=="Male"
emp[cond_gender]
```

```python
#get all rows where Sr. managment value is True
cond_sr_mangmt=emp["Senior Management"]
emp[cond_sr_mangmt]
```

```python
# all employee not in Marketing Team
cond_team=emp["Team"]!="Marketing"
emp[cond_team]
```

```python
# all employee having Salary greater than 110000
cond_sal=emp["Salary"]>110000
emp[cond_sal]
```

```python
# all employee having bonus less than 1.5
cond_bonus=emp["Bonus %"]<1.5
emp[cond_bonus]
```

```python
# all employee  who had joined on or before 01 Jan 1985
cond_st_dt=emp["Start Date"]<="1980-08-12"
emp[cond_st_dt]
```

# 3.Filter with More than One Condition (AND - &amp;)

```python
# get all Male employees having salary greater than 1,10,000
emp[cond_gender & cond_sal]
```

# 4.Filter with More than One Condition (OR - _)

```python
# get all Male employees or employees having salary greater than 1,10,000
emp[cond_gender | cond_sal]
```

```python
# get all employees whose Name is Jerry having salary greater than 1,10,000 or all employee having bonus less than 1.5
cond_name=emp["First Name"]=="Jerry"
emp[(cond_name & cond_sal)|cond_bonus].sort_values(["First Name"],ascending=[True])
# emp
```

# 7.The .between() Method

```python
# Get all employees having salary between 60000 and 70000
emp[emp["Salary"].between(60000,70000)]
```

```python
# Get all employees having salary between 60000 and 70000
emp[emp["Start Date"].between("2008-06-01","2009-12-01") & emp["Last Login Time"].between("08:30AM","12:00PM")]
```

# 8.The .duplicated() Method

```python
emp[emp["First Name"].duplicated(keep="first")].sort_values(["First Name"]) #keep the first record and consider other records as duplicate 
```

```python
emp[emp["First Name"].duplicated(keep="last")].sort_values(["First Name"]) #keep the last record and consider the other records as duplicate
```

```python
#To get all non duplicate records 
mask= ~emp["First Name"].duplicated(keep=False)
emp[mask].sort_values(["First Name"])
```

# 9.The .drop_duplicates() Method

```python
emp.drop_duplicates(subset=["First Name"],keep="first") #keep the first record and delete all the other duplicate records 
```

```python
emp.drop_duplicates(subset=["First Name"],keep="last")#keep the last record and delete all the other duplicate records
```
